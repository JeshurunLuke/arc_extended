# Generating Keithely Voltage Corrections

```python
#%matplotlib ipympl


import matplotlib.pyplot as plt  # Import library for direct plotting functions
import numpy as np  # Import Numerical Python
from IPython.display import display, HTML  # Import HTML for formatting output
import pandas as pd
from scipy.spatial.transform import Rotation as R
from .arc import *  # Import ARC (Alkali Rydberg Calculataor)

import pickle


calc = StarkMap(Rubidium87())
calc.set_cache()
calc.defineBasis(98, 2, 2.5, 2.5, 95, 101, 10,  compressedBasis=False)
```

```python
# Change this vector. These are the Keitheley voltages in which you generate corrections
# Requires a reference point. Here for state 5, we use 65 mV as reference point so its concatenated into the choice
Ectrl_ref = np.concatenate(([65], np.arange(95, 122,  2)))



# Code below converts Keitheley Voltages to real Fields
angle_scan_theta = np.zeros(len(Ectrl_ref))
angle_scan_phi = np.zeros(len(Ectrl_ref))
angle_psi = 55*np.pi/180 # Rotation angle of Eres about the y axis (kinda arbritrary)


E_scan_total = np.zeros(len(Ectrl_ref))
angle_res = 22 # Angle of res electric field
alpha = 0.73 # scaling factor for keithely voltage
Eres = 100 # Magnitude of res electric field

print(len(Ectrl_ref))
for ind, E_i in enumerate(Ectrl_ref): 
    
    ycomp = Eres*np.cos(angle_res*np.pi/180)
    zcomp = Eres*np.sin(angle_res*np.pi/180)

    
    E_res_vec = np.array([0, ycomp, zcomp])
    
    rot_axis = np.array([0, 1, 0])
    r = R.from_rotvec(angle_psi * rot_axis)
    E_res_vec = r.as_matrix() @ E_res_vec

    E_Total = E_res_vec -  np.array([0, E_i*alpha, 0])
    
    angle_scan_theta[ind] = np.arccos(np.dot(np.array([0,0, 1]), E_Total)/np.linalg.norm(E_Total))
    proj_plane = np.concatenate((np.array(E_Total[0:2]),[0]))
    proj_plane = proj_plane/np.linalg.norm(proj_plane)
    angle_scan_phi[ind] = np.arccos(np.dot(np.array([1,0, 0]), proj_plane)/np.linalg.norm(proj_plane))

    E_scan_total[ind] = np.linalg.norm(E_Total)/10
```

```python
import pickle
# Diagonalizes hamiltonian for each E-Field Point
offset = {tuple([99, 1, 3/2]): 14}

calcCorr = StarkMap(Rubidium87())
calcCorr.set_cache()
calcCorr.defineBasis(98, 2, 2.5, 0.5, 95, 101, 10,  compressedBasis=False, offset = offset)
# Magnetic Field specified here
calcCorr.diagonalise(Earray = E_scan_total,Barray = [2e-4], theta=angle_scan_theta,phi = angle_scan_phi,  drivingFromState=[0, 1, 1.5, 1.5,1])
```

```python
RES_mol = 2.227945 
RETcond = RET(calcCorr, RES_mol)    

# This is where you specify which state you want to correct for.
# THIS COUNTS BOTTOM UP. The top most state is state 10, state OI 9 (10- 1). the bottom most state in the top manifold is state 5 with respect to the bottom so stateOI = 4
stateOI = 4
refE  = 65
Es = -1*RETcond.find_energy_state(98, 2, level = stateOI)
```

```python
refE = 65 # Reference keithely voltage where correction is 0
# Correction specfied in corrOI vector
corrOI = (Es - Es[np.argmin(abs(Ectrl_ref - refE))])*1000
print(corrOI)
# Corresponding Keithely Voltage
print(Ectrl_ref)
```

# Generating expected location of resonances

```python
%matplotlib ipympl


import matplotlib.pyplot as plt  # Import library for direct plotting functions
import numpy as np  # Import Numerical Python
from IPython.display import display, HTML  # Import HTML for formatting output
import pandas as pd
from scipy.spatial.transform import Rotation as R
from arc import *  # Import ARC (Alkali Rydberg Calculataor)

import pickle


calc = StarkMap(Rubidium87())
calc.set_cache()
calc.defineBasis(98, 2, 2.5, 0.5, 95, 101, 10,  compressedBasis=False)
```

```python
# Keitheley range of Interest. 140 points can take up to 5 minutes
Ectrl_s = np.linspace(5, 100, 140)

# Code below converts Keitheley Voltages to real Fields
angle_scan_theta = np.zeros(len(Ectrl_s))
angle_scan_phi = np.zeros(len(Ectrl_s))
angle_psi = 55*np.pi/180 # Rotation angle of Eres about the y axis (kinda arbritrary)


E_scan_total = np.zeros(len(Ectrl_s))
angle_res = 22 # Angle of res electric field
alpha = 0.73 # scaling factor for keithely voltage
Eres = 104 # Magnitude of res electric field


for ind, E_i in enumerate(Ectrl_s): 
    
    ycomp = Eres*np.cos(angle_res*np.pi/180)
    zcomp = Eres*np.sin(angle_res*np.pi/180)

    
    E_res_vec = np.array([0, ycomp, zcomp])
    rot_axis = np.array([0, 1, 0])
    r = R.from_rotvec(angle_psi * rot_axis)
    E_res_vec = r.as_matrix() @ E_res_vec

    E_Total = E_res_vec -  np.array([0, E_i*alpha, 0])
    
    angle_scan_theta[ind] = np.arccos(np.dot(np.array([0,0, 1]), E_Total)/np.linalg.norm(E_Total))
    proj_plane = np.concatenate((np.array(E_Total[0:2]),[0]))
    proj_plane = proj_plane/np.linalg.norm(proj_plane)
    angle_scan_phi[ind] = np.arccos(np.dot(np.array([1,0, 0]), proj_plane)/np.linalg.norm(proj_plane))

    E_scan_total[ind] = np.linalg.norm(E_Total)/10
```

```python
import pickle

offset = {tuple([99, 1, 3/2]): 14} # Uses 14 MHz correction for quantum Defect

calc = StarkMap(Rubidium87())
calc.set_cache()
calc.defineBasis(98, 2, 2.5, 0.5, 95, 101, 10,  compressedBasis=False, offset = offset)
calc.diagonalise(Earray = E_scan_total,Barray = [2e-4], theta=angle_scan_theta,phi = angle_scan_phi,  drivingFromState=[0, 1, 1.5, 1.5,1])
```

```python
# KRb rotational spacing that sets resonance condition
RES_mol = 2.227945 
RETcond = RET(calc, RES_mol)    
sols = RETcond.find_resonance(98, 2)
```

```python
plt.clf()
convFac = 1e6
debugOutput = False
eOI = 95
# THIS COUNTS TOP DOWN
indOI = [0, 1, 2, 3, 4, 5] # this is the stateOI. The highest state is state 0 and you count down. For example, our lowest state in the manifold is state 6, so indOI, n - 1 = 5
# you can choose as many indOI as you would like
for iters, key in enumerate(reversed(sols)): 
    if iters in indOI:
        comp, e_ans = sols[key]
        if debugOutput: 
            print(comp[len(comp) -1])
            print(e_ans[0, : ])
            print(e_ans[1, : ]*convFac)
            print(Ectrl[np.argmin(np.abs(e_ans[1, : ]*convFac))])
            print(np.argmin(np.abs(e_ans[1, : ]*convFac)))
        plt.plot(Ectrl_s, (e_ans[2, : ])*convFac)
        

    
plt.show()
```

# Single E-Field and B-Field Calculation

```python
offset = {tuple([99, 1, 3/2]): 14} # Quantum defect Correction


Ectrl = 76 #Keithely voltage selected
Eres = 104
angle_res  = 22
alpha = 0.73
angle_psi = 55*np.pi/180

ycomp = Eres*np.cos(angle_res*np.pi/180)
zcomp = Eres*np.sin(angle_res*np.pi/180)
E_res_vec = np.array([0, ycomp, zcomp])
rot_axis = np.array([0, 1, 0])
r = R.from_rotvec(angle_psi * rot_axis)
E_res_vec = r.as_matrix() @ E_res_vec
E_Total = E_res_vec -  np.array([0, Ectrl*alpha, 0])
angle_theta = np.arccos(np.dot(np.array([0,0, 1]), E_Total)/np.linalg.norm(E_Total))
proj_plane = np.concatenate((np.array(E_Total[0:2]),[0]))
proj_plane = proj_plane/np.linalg.norm(proj_plane)
angle_phi = np.arccos(np.dot(np.array([1,0, 0]), proj_plane)/np.linalg.norm(proj_plane))
E_calc = np.linalg.norm(E_Total)/10

calcSet = StarkMap(Rubidium87())
calcSet.set_cache()
calcSet.defineBasis(98, 2, 2.5, 0.5, 95, 101, 10,  compressedBasis=False,  offset = offset)

# B-Field selected
calcSet.diagonalise(Earray = E_calc,Barray = [2e-4], theta=angle_theta,phi = angle_phi,  drivingFromState=[0, 1, 1.5, 1.5,1])
df0 = calcSet.displayLevels(units='ghz', displayAll = False)
df0.head(22)
```
